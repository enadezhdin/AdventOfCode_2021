```python
import sys
import os
import csv
import numpy as np
```

```python
with open ("input.txt", "r", newline = '\n') as inp_data:
    data = csv.reader(inp_data, delimiter = '\n')
    input_data = []
    for line in data:
        input_data.extend(line)
```

```python
for i in range(len(input_data)):
    input_data[i] = (",".join(input_data[i].split(' -> '))).split(",")
```

```python
input_data_clean = []
for i in range(len(input_data)):
    input_data_clean.append(list(map(int, input_data[i])))
```

```python
straight_lines = []
for i in (range(len(input_data_clean))):
    if input_data_clean[i][0] == input_data_clean[i][2] or input_data_clean[i][1] == input_data_clean[i][3]:
        #print(input_data_clean[i][0])
        straight_lines.append(input_data_clean[i])
```

```python
straight_lines_ranges = []
for i in range(len(straight_lines)):
    x_coords = [straight_lines[i][0], straight_lines[i][2]]
    y_coords = [straight_lines[i][1], straight_lines[i][3]]
    x_coords.sort()
    y_coords.sort()
    straight_lines_ranges.append([range(x_coords[0],x_coords[1]+1), range(y_coords[0],y_coords[1]+1)])
```

```python
min_number = 999
for line in input_data_clean:
    running_min = min(line)
    if running_min < min_number:
        min_number = running_min
print(min_number)
```

```python
max_number = 0
for line in input_data_clean:
    running_max = max(line)
    if running_max > max_number:
        max_number = running_max
print(max_number)
```

```python
## ok we have a matrix from 10 to 989
```

```python
diagram = np.zeros((999, 999))
```

```python
for line in straight_lines_ranges:
    for i in range(diagram.shape[0]):
        for j in range(diagram.shape[1]):
            if i in line[1] and j in line[0]:
                diagram[i,j] = diagram[i,j] + 1
```

```python
points = 0
for i in range(diagram.shape[0]):
    for j in range(diagram.shape[1]):
        if diagram[i,j] > 1:
            points = points +1
print(points)            
```

```python
import csv
import numpy as np
```

```python
with open ("input.txt", "r", newline = '\n') as inp_data:
    data = csv.reader(inp_data, delimiter = '\n')
    input_data = []
    for line in data:
        input_data.extend(line)
```

```python
draws = input_data[0]
```

```python
draws = list(map(int, draws.split(",")))
```

```python
draws
```

```python
boards = {}
k = 1
for i in range(1, len(input_data),5):
    running_list = input_data[i:i+5]
    running_matrix = []
    for row in running_list:
        running_matrix.append(list(map(int, row.split())))
    boards[f"board_{k}"] = [np.asarray(running_matrix), np.ones([5,5])]
    k += 1
```

```python
board_keys = list(boards.keys())
```

```python
collect_keys = []
win_boards = []
for num in draws:
    for key in board_keys:
        if key in win_boards:
            continue
        if num in boards[key][0]:
            coords_raw = np.where(boards[key][0] == num)
            coords = list(zip(coords_raw[0], coords_raw[1]))[0]
            boards[key][1][coords] = 0
            if 0 in np.ndarray.sum(boards[key][1], axis = 0): #
                collect_keys.append(key)
                tmp_matrix = boards[key][1]
                print(tmp_matrix)
                print(key, num)
                running_sum = np.sum(np.multiply(boards[key][0], boards[key][1]))
                print(running_sum * num)
                win_boards.append(key)
                #board_keys.remove(key) #not goood for dictionaries!
            elif 0 in np.ndarray.sum(boards[key][1], axis = 1):
                collect_keys.append(key)
                tmp_matrix = boards[key][1]
                print(tmp_matrix)
                print(key, num)
                running_sum = np.sum(np.multiply(boards[key][0], boards[key][1]))
                print(running_sum * num)
                win_boards.append(key)
                #board_keys.remove(key)
```

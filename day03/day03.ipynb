```python
import sys
import os
import csv
```

```python
with open ("input_day3.txt", "r", newline = '\n') as inp_data:
    data = csv.reader(inp_data, delimiter = '\n')
    input_data = []
    for line in data:
        input_data.extend(line)
#input_data = list(map(int, input_data))
```

```python
count_dict = {}
for i in range(len(input_data[0])):
    running_reg = 0
    for j in range(len(input_data)):
        if input_data[j][i] == '1':
            running_reg +=1
    count_dict[f'reg_{i}'] = running_reg
```

```python
count_dict
```

```python
max_number = []
for val in count_dict.values():
    if val/len(input_data) > 0.5:
        max_number.extend('1')
    else:
        max_number.extend('0')
print(max_number)
```

```python
max_number_num = "".join(max_number)
```

```python
min_number = max_number[:]
for i, digit in enumerate(max_number):
    if digit == '1':
        min_number[i] = '0'
    else:
        min_number[i] = '1'
print(min_number)
```

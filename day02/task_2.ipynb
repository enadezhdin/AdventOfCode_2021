```python
import sys
import os
import csv
```

```python
with open ("input.txt", "r", newline = '\n') as inp_data:
    #data = inp_data.read()
    data = csv.reader(inp_data, delimiter = '\n')
    input_data = []
    for line in data:
        input_data.extend(line)
#input_data = list(map(int, input_data))
```

```python
input_data[3:10]
```

```python
input_data[3].split(' ')[1]
```

```python
current_x = 0
current_y = 0
for i in range(len(input_data)):
    direct = input_data[i].split(' ')[0]
    value = int(input_data[i].split(' ')[1])
    if direct == 'forward':
        current_x = current_x + value
    if direct == 'up':
        current_y = current_y - value
    if direct == 'down':
        current_y = current_y + value
final_val = current_x * current_y
```
